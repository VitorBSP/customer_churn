```python
import random
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('data/new_customer_nps.csv')
```

```python
df
```

```python
output_path = 'data/customer_nps.csv'

df_nps = df[['cpf', 'Predicted_NPS']].copy()

df_nps.rename(columns={'Predicted_NPS': 'NPS'}, inplace=True)

df_nps['cpf'] = (
    df_nps['cpf']
      .astype(str)
      .str.replace(r'\D', '', regex=True)
)

df_nps.to_csv(output_path, index=False)
```

```python
output_path = 'data/customer_social.csv'

df_social = df[['cpf', 'gender', 'SeniorCitizen', 'Partner', 'Dependents']].copy()

df_social['cpf'] = (
    df_social['cpf']
      .astype(str)
      .str.replace(r'\D', '', regex=True)
      .astype('Int64')                # inteiro com suporte a NA
)

df_social.to_csv(output_path, index=False)
```

```python
output_path = 'data/customer_original.csv'
df.drop(['Predicted_NPS', 'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'customerID', 'RandomFactor', 'Churn_flag'], axis=1).to_csv(output_path, index=False)
```
